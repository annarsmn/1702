```python
def compress(text: list[str]) -> int:
  """На вход подаются только большие английские буквы. Если подряд больше одной буквы то сдимем до записи <Буква><Количество> - возвращаем новую длину массива символов. Нельзя использовать доп память"""

  if not text:
    return 0

  if len(text) == 1:
    return 1

  for x in text:
    if ord(x) not in range(65,91):
      return 0

  slow=fast=1

  while fast<len(text):
    quantity = 1
    if text[fast] == text[fast-1]:
      while fast<len(text) and text[fast] == text[fast-1]:
        quantity += 1
        fast += 1
      if quantity != 1:
        str_quantity = str(quantity)
        text[slow:slow+len(str_quantity)] = str_quantity
        slow += len(str_quantity)
    if fast == len(text)-1:
      text[slow] = text[fast]
      slow+=1
      return slow
    if fast<len(text) and slow<len(text):
      text[slow]=text[fast]
      slow+=1
    fast+=1
  return slow


text = list('AAAABBBCCXYZDDDDEEEFFFAAAAAABBBBB')

new_len = compress(text)
assert new_len == 19
assert text[:new_len] == list('A4B3C2XYZD4E3F3A6B5')

text = list('AAAABBBCCXYZDDDDEEEFFFAAAAAABBBBBBBBBBBBBBBBBBBBBBBBBBBB')

new_len = compress(text)
assert new_len == 20
assert text[:new_len] == list('A4B3C2XYZD4E3F3A6B28')


text = list('ABC')
new_len = compress(text)
assert new_len == 3
assert text[:new_len] == list('ABC')

text = list('ABBBCERTEEE')
new_len = compress(text)
assert new_len == 9
assert text[:new_len] == list('AB3CERTE3')

text = list('AAAA')
new_len = compress(text)
assert new_len == 2
assert text[:new_len] == list('A4')

text = list('')
new_len = compress(text)
assert new_len == 0
assert text[:new_len] == list('')

text = list('abs')
new_len = compress(text)
assert new_len == 0
assert text[:new_len] == list('')

text = list('Ab')
new_len = compress(text)
assert new_len == 0
assert text[:new_len] == list('')

text = list('A1')
new_len = compress(text)
assert new_len == 0
assert text[:new_len] == list('')
```
