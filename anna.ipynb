```python
def urlify(text: list[str]) -> int:
  """Заменяем пробелы на %20 и возвращаем новую длину массива символов"""
  spaces = 0
  len1 = 0

  for x in text:
    if x!='':
      len1 += 1
#считаю количество непустых элементов, чтобы отделить данную строку от дополнительной памяти
    if x ==' ':
      spaces += 1

  len2 = len1+spaces*2
  if len2 == len1:
    return len1

  new = len2-1
  old = len1-1

  while old >=0:
    if text[old] == ' ':
      text[new] = '0'
      text[new-1] = '2'
      text[new-2] = '%'
      new = new - 3
    else:
      text[new] = text[old]
      new -= 1
    old -=1

  return len2

text = list('my url') + [''] * 21
new_len = urlify(text)

assert new_len == 8
assert text[:new_len] == list('my%20url')

text = list(' my  url ') + [''] * 21
new_len = urlify(text)

assert new_len == 17
assert text[:new_len] == list('%20my%20%20url%20')

text = list('') + [''] * 21
new_len = urlify(text)

assert new_len == 0
assert text[:new_len] == list('')

text = list('  ') + [''] * 21
new_len = urlify(text)

assert new_len == 6
assert text[:new_len] == list('%20%20')

text = list(' ') + [''] * 21
new_len = urlify(text)

assert new_len == 3
assert text[:new_len] == list('%20')

text = list('%20') + [''] * 21
new_len = urlify(text)

assert new_len == 3
assert text[:new_len] == list('%20')

text = list('text') + [''] * 21
new_len = urlify(text)

assert new_len == 4
assert text[:new_len] == list('text')
```

```python
def remove_zeroes(numbers: list[int]) -> int:
  """Переносим все 0 конец и возвращаем индекс первого нуля"""
  slow = fast = 0

  while fast < len(numbers):
      if numbers[fast] != 0:
        numbers[slow], numbers[fast] = numbers[fast], numbers[slow]
        slow +=1
      fast +=1

  return slow


nums = [0,42,21,0,100,0,5,1,0,7,3,0,404,0]
new_len = remove_zeroes(nums)

assert new_len == 8
assert nums[:new_len] == [42,21,100,5,1,7,3,404]

nums = [0,0,0,0]
new_len = remove_zeroes(nums)

assert new_len == 0
assert nums[:new_len] == []

nums = []
new_len = remove_zeroes(nums)

assert new_len == 0
assert nums[:new_len] == []

nums = [1,2,3]
new_len = remove_zeroes(nums)

assert new_len == 3
assert nums[:new_len] == [1,2,3]

nums = [0]
new_len = remove_zeroes(nums)

assert new_len == 0
assert nums[:new_len] == []

nums = [1,0,2,3,0]
new_len = remove_zeroes(nums)

assert new_len == 3
assert nums[:new_len] == [1,2,3]
```

```python
def compress(text: list[str]) -> int:
  """На вход подаются только большие английские буквы. Если подряд больше одной буквы то сдимем до записи <Буква><Количество> - возвращаем новую длину массива символов. Нельзя использовать доп память"""

  if text=='':
    return None

  slow=fast=1

  while fast<len(text):
    quantity = 1
    if text[fast]==text[fast-1]:
      while fast<len(text) and text[fast] == text[fast-1]:
        quantity += 1
        fast += 1
      if quantity !=1:
        text[slow] = str(quantity)
        slow +=1
    if slow<len(text) and fast<len(text):
      text[slow]=text[fast]
      slow+=1
      fast+=1
  return slow

text = list('AAAABBBCCXYZDDDDEEEFFFAAAAAABBBBBBBBBBBBBBBBBBBBBBBBBBBB')

new_len = compress(text)
print(compress(text))
print(text)
assert new_len == 20
assert text[:new_len] == list('A4B3C2XYZD4E3F3A6B28')
```
